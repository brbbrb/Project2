```python
#import data from CSV file
covid_deaths_df = pd.read_csv('united_states_covid19_cases_deaths_and_testing_by_state.csv')

covid_deaths_df
```

```python
# Rename "New York*" to "New York"
covid_deaths_df['State/Territory'] = covid_deaths_df['State/Territory'].replace(['New York*'],'New York')
```

```python
# Create US States list
states_list = 'Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi', 'Missouri', 'Montana', 'Nebraska', 'Nevada', 'New Hampshire', 'New Jersey', 'New Mexico', 'New York', 'North Carolina', 'North Dakota', 'Ohio', 'Oklahoma', 'Oregon', 'Pennsylvania', 'Rhode Island', 'South Carolina', 'South Dakota', 'Tennessee', 'Texas', 'Utah', 'Vermont', 'Virginia', 'Washington', 'West Virginia', 'Wisconsin', 'Wyoming'
```

```python
# Remove values in "State/Territory" column that are not the 50 US States
covid_deaths_states = covid_deaths_df[covid_deaths_df['State/Territory'].isin(states_list)]

# Rename "State/Territory" column to "State"
covid_deaths_states.rename(columns={'State/Territory':'State'}, inplace=True)

# Reset the index
covid_deaths_states.reset_index(drop=True, inplace=True)
```

```python
# Create a new dataframe with only the columns we need to answer our questions
covid_deaths_st_need = covid_deaths_states[['State','Total Cases','Total Deaths']]
covid_deaths_st_need.head(10)
```
